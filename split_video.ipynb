```python
vid = cv2.VideoCapture("carsa_trim.mp4") #change 
vid.set(cv2.CAP_PROP_POS_FRAMES, 299)
success,image = vid.read()
count = 0

try:
    if not os.path.exists("carsa_trim_folder"):
        os.makedirs("carsa_trim_folder")
except OSError:
    print("Error: Creating directory of data")
    
while success:
    # image = image[:,100:1920-100]
    # for some reason some of the videos are rotated when opened
#     image = cv2.rotate(image, cv2.ROTATE_180)
    cv2.imwrite("carsa_trim_folder/carsa_trim_real_%d.jpg" % count, image)     # save frame as JPEG file   

    success,image = vid.read()
    print('Read frame : ', success)
#     if count > 10:
#       success = False
    count += 1
```
